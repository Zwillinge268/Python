## 格式化輸出

---
%d = int , %f = float , %s = string

%5d = 固定5位數，不足則補空格

%-5d = 固定5位數，不足則補空格(補在後方)

%8.2f = 固定8字元以及2位小數(字元包括小數點)，不足則補空格

```python
price,temp = 23.8,"價格 = "
print("%10s %8.2f" %(temp, price))
```

##for迴圈 與 range

```python
list1 = list(range(5))
print(list1)

list2 = list(range(3,8,2))     #range(起始，中止，間隔) if(中止 > 起始) 間隔一定為正數
print(list2)

list3 = list(range(8,3,-1))    #range(起始，中止，間隔) if(起始 > 中止) 間隔一定為負數
print(list3)
```

```python
#for 變數 in 陣列 (陣列有多少東西就執行多少遍，每次執行將該陣列元素賦予變數)
list1 = ["A","B","C"]
for i in list1:
  print(i)

#以 range 產生陣列
for j in range(3,8,2):
  print(j,end='')
```

### 雞兔同籠，若共有35個頭，94隻腳，請問有幾隻雞?幾隻兔?

```python
for chicken in range(0,36):
    rabbit=35-chicken
    if (2*chicken+4*rabbit) == 94:
        print("雞有",chicken,"隻,兔子有",rabbit,"隻")
        break
else:
    print("無解")
```

## while迴圈

```python
sum=0
count=1
while count <=10:
    sum += count
    count += 1
print("1+2+3+...+10=",sum)
```

### 本金10000元，定存年利率0.6%，請問多久銀行存款會翻倍

```python
x=10000
years=0
while x <= 20000:
    x=x*1.006
    years += 1
print(str(years), "年以後，存款加倍，為", int(x),"元")
```

##基本模組 - Random

```python
import random
print(random.randint(1,10))   #隨機產生n1(含)到n2(含)之間的整數
print(random.uniform(1,10))   #隨機產生n1到n2之間的浮點數
print(random.random())     #隨機產生一個0到1之間的浮點數
```

```python
import random
list1 = [10,30,50,70,90,110,130,150,170,190,210,230,250,270,290]

for i in range(5):
  print(random.choice(list1))   #隨機從某陣列/元組中抽取值
```
